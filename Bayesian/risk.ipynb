برای طبقه‌بندی بیزین، ما با داده‌ها یا الگوهایی سر و کار داریم که می‌خواهیم تشخیص دهیم به کدام دسته یا کلاس تعلق دارند. مثلا، فرض کنید می‌خواهیم تشخیص دهیم ایمیل‌ها اسپم هستند یا نه.

برای اینکار با مفاهیم زیر باید اشنا بود:

**:احتمال پیشین**

احتمال پیشین فرض اولیه ما درباره احتمال تعلق یک داده به کلاسی است. قبل از اینکه به داده‌ها نگاه کنیم، یک حدسی درباره‌ی احتمال اینکه یک الگو به یک کلاس خاص تعلق داره، داریم. مثلا، فکر می‌کنیم ۵۰ درصد ایمیل‌ها اسپم هستند.



**:تابع چگالی احتمال**

تابع چگالی احتمال نشان‌دهنده توزیع احتمالی ویژگی‌های داده‌ها در هر کلاس است. به عنوان مثال،  وقتی داده‌ای مثل یک ایمیل رو داریم، با توجه به ویژگی‌هایی مثل کلمات موجود در اون، احتمال اینکه اسپم باشه چقدره.


**:احتمال پسین**

 احتمال پسین احتمال به‌روز شده پس از دیدن داده است. با استفاده از قانون بیز، این دو اطلاعات (احتمال پیشین و تابع چگالی احتمال) رو با هم ترکیب می‌کنیم تا بفهمیم بعد از دیدن داده‌ها، احتمال اینکه هر الگو به کدام کلاس تعلق داره، چقدره. مثلا، بعد از خواندن یک ایمیل، تصمیم می‌گیریم که احتمال اینکه اسپم باشه ۸۰ درصده. و در نهایت، بر اساس احتمال پسینی که به دست آوردیم، تصمیم می‌گیریم که هر الگو رو به کدوم کلاس نسبت بدهیم. اگر احتمال اسپم بودن ایمیل بیشتر از حد معینی باشه، می‌گوییم ایمیل اسپم است.

## example

class normal: alpha=16 betta=8

class fault: alpha=3 betta=4

image.png

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

x = np.linspace(0, 5, 1000)
A = stats.gamma.pdf(x, a=16, scale=1/8)
B = stats.gamma.pdf(x, a=3, scale=1/4)
min = np.minimum(A, B)

plt.plot(x, A, color='red', label='A')
plt.plot(x, B, color='blue', label='B')
plt.fill_between(x, min, color='green', alpha=0.5)
plt.legend()
plt.show()
```
